```python
#open, decode as utf-u and read the file
myfile = open("englishsongs.txt", 'r', encoding='utf‚Äê8')
songs = myfile.read()
```

```python

print(len(songs)) #check the length of the data I am working with
```

```python
word_tokenized_songs = songs.split(' ') # tokenise the text in strings
```

```python
word_tokenized_songs = [w.lower() for w in word_tokenized_songs] #lowercase the text to normalise it, which prevents two words as 'I' and 'i' to be treated differently
print (word_tokenized_songs)
```

```python
cleaning_tokens = [] #creates an empty list to store words without \n
for item in word_tokenized_songs:
    # replace \n with space
    item = item.replace('\n', ' ')

    # split on any whitespace (spaces, tabs, etc.)
    words = item.split()

    # add all split words to final list
    cleaning_tokens.extend(words)
```

```python
print (cleaning_tokens)
```

```python
# here i wanted to split contractions having a subject, but i didn't want to split negations as in don't or wouldn't
subject_contractions = {
    "i'm": ["i", "am"],
    "i'll": ["i", "will"],
    "i'd": ["i", "would"],
    "i've": ["i", "have"],
    "you're": ["you", "are"],
    "you'll": ["you", "will"],
    "you'd": ["you", "would"],
    "you've": ["you", "have"],
    "he's": ["he", "is"],
    "she's": ["she", "is"],
    "it's": ["it", "is"],
    "we're": ["we", "are"],
    "we'll": ["we", "will"],
    "we'd": ["we", "would"],
    "we've": ["we", "have"],
    "they're": ["they", "are"],
    "they'll": ["they", "will"],
    "they'd": ["they", "would"],
    "they've": ["they", "have"],
}
normalising_tokens = []
for word in cleaning_tokens:
    # check if word is a subject contraction
    if word in subject_contractions:
        # split into separate words
        normalising_tokens.extend(subject_contractions[word])
    else:
        # keep as is (including "don't", "can't", "won't", "isn't", etc.)
        normalising_tokens.append(word)

print (normalising_tokens)
```

```python
from string import punctuation #import punctuation

final_clean_words = [] #creates an empty list to store cleaned words
for word in normalising_tokens:
    clean_word = word # creates a copy of the original word to modify
    # Remove each punctuation character
    for punc in punctuation:
        clean_word = clean_word.replace(punc, '')
    if clean_word:
        final_clean_words.append(clean_word) #check if strings are not empty

print (final_clean_words)
```

```python
from collections import Counter #import counter class for frequency counting

freq_songs = Counter(final_clean_words) # create frequency distribution of cleaned words
top_15 = freq_songs.most_common(15) # get the 15 most common words with their frequency

print (freq_songs.most_common(15))
```

```python
top_15_with_lengths = [] #create list to store words with their frequency and length
for word, freq in top_15: #unpack each tuple: (word, frequency)
    word_length = len(word) #calculate character count of the word
    top_15_with_lengths.append((word, freq, word_length)) #store as triple

print (top_15_with_lengths)
```

```python
#display the top 15 words as a formatted table
print("=" * 40) #print top border line
print("RANK  WORD        FREQ  LENGTH") #column headers
print("-" * 40) # header separator
for i, (word, freq, length) in enumerate(top_15_with_lengths, 1):  # i=rank and start counting the rank from 1 onwards
    print(f"{i:4}  {word:10}  {freq:4}  {length:6}") #format each row with fixed-width columns for alignment
print("=" * 40) #print bottom border line
```
